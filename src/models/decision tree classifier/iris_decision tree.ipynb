# Decision Tree

## 1. Import the needed Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2.	Import the tools needed in Sklearn

```python

import sklearn
from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.tree import (DecisionTreeClassifier,plot_tree)
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

## 3. Call the data

```python
iris=pd.read_csv("../../data/Iris.csv")
iris
```

## 4. Define Feature and  Target

```python
feature=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
target=['Species']
x=iris[feature]
y=iris[target]
```

## 5.	Split dataset into training and validation set

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.25, random_state=42)
x_train
```

## 6.	Train Decision Tree Model 

(you can check the meaning of each parameter here: https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html )

```python
dt_model=DecisionTreeClassifier(criterion='entropy', max_depth=2)
dt_model.fit(x_train,y_train)
```

## 7.	Test Model

```python
print("Train data accuracy:", accuracy_score(y_true=y_train, y_pred=dt_model.predict(x_train)))
print("Test data accuracy:", accuracy_score(y_true=y_test, y_pred=dt_model.predict(x_test)))
```

## 8.	Plot Decision Tree 

```python
plt.figure(figsize=(10,10))
plot_tree(dt_model, filled=True, class_names=True)
plt.title("Iris Decision Tree Classifier")
plt.show()

plt.savefig('../../visualizations/iris_decision tree.png')
```

## 9.	Predict New Dataset

```python
x_new=[[5.9,3.0,5.1,1.8]]
print(dt_model.predict(x_new))
print(dt_model.predict_proba(x_new))
```
