```python
import matplotlib.pyplot as plt
```

```python
# plot the histogram of the sepal length
plt.hist(iris['SepalLengthCm'], bins=20)
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# plot a line graph of the sepal length, sepal width, petal length, and petal width
plt.plot(iris['SepalLengthCm'], label='Sepal Length')
plt.plot(iris['SepalWidthCm'], label='Sepal Width')
plt.plot(iris['PetalLengthCm'], label='Petal Length')
plt.plot(iris['PetalWidthCm'], label='Petal Width')
plt.title('Line Graph of Iris Measurements')
plt.xlabel('Observation')
plt.ylabel('Measurement (cm)')
plt.legend()
plt.show()
```
